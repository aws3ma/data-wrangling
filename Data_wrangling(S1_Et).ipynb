* load dataset

```python
url = "https://github.com/byuidatascience/data4python4ds/raw/master/data-raw/who/who.csv"
db=pd.read_csv(url)
```

* Garder seulement les 17 premières colonnes

```python
df1 = db.iloc[:,0:18]
df1
```

* Supprimer les observations contenant la valeur NaN

```python
df1=df1.dropna(axis=0)
df1
```

* Transformer les variables new_sp_???? en des valeurs 

```python
df1 = df1.melt(
    id_vars=['country','iso2','iso3','year'],
    var_name = 'key',
    value_name='cases'
)
df1
```

* Supprimer de chaque valeur de la colonne key le préfixe new_sp_ 

```python
df1 = df1.assign(key = lambda x : x.key.str.replace('new_sp_',''))
df1
```

* Séparer la colonne Key en deux colonnes: Genre (M, F) et Age 

```python
df2 = df1.assign(
    sex=df1.key.str[:1],
    age=df1.key.str.slice(1)
)
df2
```

* Drop iso2, iso3 and key columns 

```python
df3 = df2.filter(items=['country','year','sex','age','cases'])
df3
```
